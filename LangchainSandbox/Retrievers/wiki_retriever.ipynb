### rich Console

```python
from rich.console import Console
from rich.pretty import Pretty
from rich.theme import Theme

theme = Theme(
    {
        "json.key": "#c1a2ff",
        "json.string": "#FCCEA1",
        "json.number": "#A1C4FD",
        "json.boolean": "#A6F5D8",
        "json.null": "#ffb3ba",
    }
)
console = Console(
    theme=theme,
    force_terminal=True,
    force_jupyter=False,
    color_system="truecolor",
)
```

### Retriever

```python
from langchain_community.retrievers import WikipediaRetriever

wiki_retriever = WikipediaRetriever(top_k_results=2, lang="en")
```

```python
res = wiki_retriever.invoke("Who is Carl Segan?")

console.print(Pretty(res))
```
