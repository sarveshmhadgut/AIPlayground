```python
from langchain_chroma import Chroma
from langchain_google_genai import GoogleGenerativeAIEmbeddings

embeddings = GoogleGenerativeAIEmbeddings(model="gemini-embedding-001")
store = Chroma.from_documents(
    documents=documents, embedding=embeddings, persist_directory="./chroma_store"
)
```

```python
console.print(Pretty(store.add_documents(documents=documents)))
```

```python
console.print(Pretty(store.get(include=["documents"])))
```

```python
query = "What is dark matter?"
```

### Similarity Based

```python
semantic_retriever = store.as_retriever(
    search_type="similarity", search_kwargs={"k": 3}
)
console.print(Pretty(semantic_retriever.invoke(query)))
```

### Maximal Marginal Relevance (MMR) Based

```python
mmr_retriever = store.as_retriever(
    search_type="mmr", search_kwargs={"k": 3, "lambda_mul": 0.5}
)
console.print(Pretty(mmr_retriever.invoke(query)))
```
