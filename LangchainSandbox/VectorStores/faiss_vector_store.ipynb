### FAISS Vector Store

```python
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
store = FAISS.from_documents(documents, embeddings)
```

### Retrieve Documents

```python
console.print(Pretty(store.docstore._dict))
```

### Retreive Document by Id

```python
console.print(Pretty(store.get_by_ids(["783e7b6e-c83f-4848-a70f-4f6b60357ae7"])))
```

### Add Documents

```python
new_doc = Document(
    page_content=(
        "The event horizon is the critical boundary surrounding a black hole where the escape velocity "
        "equals the speed of light. Once any object or radiation crosses this boundary, it cannot return "
        "to the outside universe because spacetime itself is curved inward. The radius of the event horizon "
        "is called the Schwarzschild radius, which depends directly on the mass of the black hole. "
        "The event horizon does not have a physical surface but represents a geometric limit in spacetime."
    ),
    metadata={
        "topic": "black_holes",
        "subtopic": "event_horizon",
        "difficulty": "intermediate",
        "version": 2,
        "updated": True,
        "source": "astronomy_reference",
    },
)

new_id = store.add_documents([new_doc])[0]
console.print("New IDs:", Pretty(new_id))
```

### Update Document (Emulated: delete+add)

```python
updated_doc = Document(
    page_content=(
        "The event horizon is the critical boundary surrounding a black hole where the escape velocity "
        "equals the speed of light. Once any object or radiation crosses this boundary, it cannot return "
        "to the outside universe because spacetime itself is curved inward. The radius of the event horizon "
        "is called the Schwarzschild radius, which depends directly on the mass of the black hole. "
        "The event horizon does not have a physical surface but represents a geometric limit in spacetime."
    ),
    metadata={
        "topic": "black_holes",
        "subtopic": "event_horizon",
        "difficulty": "intermediate",
        "version": 2,
        "updated": True,
        "source": "astronomy_reference",
    },
)

store.delete([new_id])

new_id = store.add_documents([updated_doc])[0]
console.print("New IDs:", Pretty(new_id))
```

### Save Locally

```python
store.save_local("faiss_store")
```

### Load Local Store

```python
store = FAISS.load_local(
    "faiss_store", embeddings=embeddings, allow_dangerous_deserialization=True
)
```

### Similarity Search

```python
console.print(Pretty(store.similarity_search(query="What is dark matter?", k=3)))
```

### Similarity Search with score

```python
console.print(Pretty(store.similarity_search_with_score("What is dark matter?", k=3)))
```

### Filtered Similarity Search

```python
console.print(
    Pretty(
        store.similarity_search_with_score(
            query="Explain Milky Way", filter={"topic": "galaxies"}, k=3
        )
    )
)
```
