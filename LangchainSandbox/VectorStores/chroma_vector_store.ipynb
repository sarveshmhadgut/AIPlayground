### Chroma Vector Store 

```python
from langchain_chroma import Chroma
from langchain_google_genai import GoogleGenerativeAIEmbeddings

embeddings = GoogleGenerativeAIEmbeddings(model="gemini-embedding-001")
store = Chroma(
    embedding_function=embeddings,
    persist_directory="./VectorStores/chroma_store",
    collection_name="astronomy",
)
```

### Update Documents

```python
updated_doc = Document(
    page_content=(
        "The event horizon is the critical boundary surrounding a black hole where the escape velocity "
        "equals the speed of light. Once any object or radiation crosses this boundary, it cannot return "
        "to the outside universe because spacetime itself is curved inward. The radius of the event horizon "
        "is called the Schwarzschild radius, which depends directly on the mass of the black hole. "
        "The event horizon does not have a physical surface but represents a geometric limit in spacetime."
    ),
    metadata={
        "topic": "black_holes",
        "subtopic": "event_horizon",
        "difficulty": "intermediate",
        "id": 5,
        "version": 2,
        "updated": True,
        "source": "astronomy_reference",
    },
)

store.update_document(
    document_id="f2dd03ae-e80b-4df0-ba78-d53486e8ade8", document=updated_doc
)

console.print(Pretty(store.get_by_ids(["f2dd03ae-e80b-4df0-ba78-d53486e8ade8"])))
```
